```python
r_dataset = pd.read_csv("/kaggle/input/groceries/groceries - groceries.csv")
r_dataset.head(3)
```

```python
type(r_dataset["Item 5"][0])
```

```python
dataset = r_dataset[[ 'Item 1', 'Item 2', 'Item 3', 'Item 4', 'Item 5', 'Item 6',
       'Item 7', 'Item 8', 'Item 9', 'Item 10', 'Item 11']]
```

```python
dataset.head(3)
```

```python
market = []
for i in range(0,dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if(type(dataset[j][i])==str):
            cus.append(dataset[j][i])
    market.append(cus)
```

```python
l = [] 
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
df = {
    "Item Name":p.keys(),
    "values":p.values()
}
```

```python
pd.set_option("display.max_rows",200)
```

```python
n_dataset = pd.DataFrame(df)
n_dataset.sort_values(by="values",ascending=True)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder 
```

```python
tr = TransactionEncoder()
```

```python
tr
```

```python
df = pd.DataFrame(tr.fit_transform(market),columns=tr.columns_)
```

```python
df
```

### Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.05,use_colnames=True,max_len=3).sort_values(by="support",ascending=True)
```
