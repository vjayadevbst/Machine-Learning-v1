# UnderSampling

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"/content/Social_Network_Ads.csv")
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[31,118000]])
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
x_u,y_u = ru.fit_resample(x_train,y_train)
```

```python
x_u.shape,y_u.shape
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x_u,y_u,test_size=0.2,random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x1_train,y1_train)
```

```python
lr1.score(x1_test,y1_test)
```

```python
lr.predict([[31,118000]])
```

```python
lr.predict([[46,41000]])
```
